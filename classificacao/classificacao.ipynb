### INTRODUÇÃO

A inteligência artificial normalmente opera com bases de dados que, apesar de extensas, são incompletas. Assim, frequentemente é necessário trabalhar com incertezas, tentando prever um futuro incerto. Por exemplo, você pode tentar prever o desempenho de um campeão olímpico durante as próximas Olimpíadas com base em resultados anteriores. Mesmo que tenham vencido antes, isso não significa que vão vencer desta vez. Fatores imprevisíveis, como uma discussão com o parceiro esta manhã ou a falta de tempo para o café da manhã, podem ou não influenciar seus resultados. No aprendizado de máquina, mais especificamente, estamos interessados em probabilidades condicionais, isto é, na probabilidade de que algo aconteça, dado que algo mais tenha acontecido anteriormente. Por exemplo, usando a probabilidade condicional, podemos tentar responder à pergunta de qual é a probabilidade de que um atleta vença uma corrida, dados seus resultados nos últimos 10 anos. Portanto, a incerteza é parte integrante da modelagem de aprendizado de máquina, pois a realidade é bem complicada e nada é perfeito. Nesse sentido, existem certos algoritmos no aprendizado de máquina que ajudam a encontrar uma maneira de contornar essa limitação.

O algoritmo de aprendizado de máquina Naive Bayes é uma das ferramentas para lidar com a incerteza com o auxílio de métodos probabilísticos. Trata-se de uma técnica simples de classificação baseado no teorema de Bayes com uma suposição de independência entre os preditores. Em termos simples, um classificador Naive Bayes assume que a presença de uma característica particular em uma classe não está relacionada com a presença de qualquer outra. Por exemplo, um fruto pode ser considerado uma maçã se é vermelho, redondo, e tiver cerca de 3 polegadas de  diâmetro. Mesmo que essas características dependam umas das outras ou da existência de outras características, todas estas propriedades contribuem de forma independente para a probabilidade de que este fruto é uma maçã e é por isso que é conhecido como "Naive" (que significa ingênuo). As suposições feitas por Naive Bayes geralmente não são corretas em situações do mundo real. Na verdade, a suposição de independência nunca é correta, mas geralmente funciona bem na prática. Para alguns tipos de modelos de probabilidade, os classificadores Naive Bayes podem ser treinados com muita eficiência em um ambiente de aprendizado supervisionado. Naive Bayes é conhecido por ganhar de métodos de classificação altamente sofisticados, além de ser simples e requerer apenas um pequeno número de dados de treinamento para estimar os parâmetros necessários para a classificação.

Antes de discutirmos como o algoritmo de Naive Bayes opera, precisamos entender rapidamente o conceito matemático por trás do método: o teorema de Bayes. O teorema de Bayes é utilizado para calcular probabilidades condicionais. É extremamente útil porque nos permite usar algum conhecimento anterior para calcular a probabilidade de um evento relacionado. É usado no desenvolvimento de modelos para classificação e problemas de modelagem preditiva. O importante é que não podemos apenas descobrir como as evidências afetam a probabilidade de um evento, mas também em quanto. Matematicamente, o teorema de Bayes é expresso como:

***P(A|B) = P(B|A) * P(A) / P(B)***

Assim, podemos encontrar a probabilidade de ***A***, dado que ***B*** ocorreu. ***A*** é a hipótese e ***B*** é a evidência. Na equação acima, temos que:

***P(B|A)*** é a probabilidade de ***B*** dado que ***A*** é verdadeiro.

***P(A)*** e ***P(B)*** são as probabilidades independentes de ***A*** e ***B***.

O classificador Naive Bayes calcula a probabilidade de um evento nas seguintes etapas:

*   Etapa 1: calcular a probabilidade anterior para determinados rótulos de classe;
*   Etapa 2: calcular a probabilidade de verossimilhança com cada atributo para cada classe;
*   Etapa 3: utilizar o teorema de Bayes para calcular a probabilidade posterior;
*   Etapa 4: analisar qual classe tem maior probabilidade, dado que a entrada pertence à classe de maior probabilidade.

Sempre que você executa a classificação, a primeira etapa é entender o problema e identificar características e rótulos potenciais. Características são aqueles atributos que afetam os resultados do rótulo. Por exemplo, no caso de uma distribuição de empréstimo, o gerente do banco identifica a ocupação, renda, idade, localização, histórico de empréstimo anterior, histórico de transações e pontuação de crédito do gerente do banco. Essas características são conhecidas como recursos que auxiliam o modelo na classificação dos clientes.

A classificação tem duas fases, a fase de aprendizagem e a fase de avaliação. Na fase de aprendizado, o classificador treina seu modelo em um determinado conjunto de dados e, na fase de avaliação, testa o desempenho do classificador. O desempenho é avaliado com base em vários parâmetros, como exatidão, erro, precisão e recuperação.

Em relação aos prós e contras do algoritmo de Naive Bayes, podemos considerar:

Prós: 

*   É fácil, rápido e muito preciso para prever o conjunto de dados da classe de teste;
*   Tem um bom desempenho na previsão de classes múltiplas;
*   Quando a suposição de independência prevalece, tem melhor desempenho em comparação com outros modelos, como a regressão logística;
*   O desempenho é bom em caso de variáveis categóricas de entrada comparada com a variáveis numéricas;
*   Para variáveis numéricas, assume-se a distribuição normal, que é uma suposição forte;
*   Tem um custo de computação muito baixo;
*   Tem um bom desempenho no caso de variável de resposta discreta em comparação com a variável contínua;
*   Trabalha com eficiência em um grande conjunto de dados.

Contras: 

*   Se a variável categórica tem uma categoria (no conjunto de dados de teste) que não foi observada no conjunto de dados de treinamento, então o modelo irá atribuir uma probabilidade de zero e não será capaz de fazer uma previsão;
*   A suposição de preditores independentes não é realmente válida. É quase impossível ter um conjunto de indicadores que sejam completamente independentes.

Considerando esse último ponto, você poderia se perguntar: "Se a hipótese de que os preditores são independentes quase nunca é válida considerando dados reais, por que utilizaria Naive Bayes?". Acontece que o algoritmo é capaz de resolver com eficácia muitos problemas complexos. Apesar de seu design ingênuo e suposições aparentemente simplificadas, esses classificadores funcionaram muito bem em muitas situações complexas do mundo real. Em 2004, uma análise do problema de classificação bayesiana mostrou que há razões teóricas sólidas para a eficácia aparentemente implausível de classificadores Naive Bayes. O modelo funciona bem mesmo com dados insuficientes ou com rótulos incorretos, então você não precisa "alimentá-lo" com centenas de milhares de exemplos antes de conseguir algo razoável com ele.

Dentre as principais aplicações do método Naive Bayes, podemos listar:

*   Classificação de documentos: determina a qual categoria um determinado documento pertence, podendo ser usado para classificar textos em diferentes idiomas, gêneros ou tópicos;
*   Filtragem de spam: classifica facilmente o spam usando palavras-chave. O algoritmo deve ser treinado para reconhecer tais probabilidades e, então, pode aplicá-las de forma eficiente para filtragem de spam;
*   Análise de sentimentos: com base nas emoções que as palavras em um texto expressam, pode calcular a probabilidade de ser positivo ou negativo. Por exemplo, em avaliações de clientes, "bom" ou "barato" geralmente significa que o cliente está satisfeito;
*   Previsão em tempo real: pode ser usado para fazer previsões em tempo real;
*   Predição multiclasse: prevê a probabilidade de várias classes de variável de destino;
*   Classificação de imagens: pode ser treinado para reconhecer dígitos escritos à mão ou colocar imagens em categorias por meio do aprendizado de máquina supervisionado;
*   Sistema de recomendação: o classificador Naive Bayes e a filtragem colaborativa juntos criam um sistema de recomendação que usa técnicas de aprendizado de máquina e mineração de dados para filtrar informações invisíveis e prever se um usuário gostaria de um determinado recurso ou não.

Resumindo, classificadores Naive Bayes podem ser extremamente rápidos em comparação com métodos mais sofisticados. O desacoplamento das distribuições de características condicionais de classe significa que cada distribuição pode ser estimada independentemente como uma distribuição unidimensional. Apesar de suas suposições aparentemente simplificadas demais, os classificadores Naive Bayes funcionaram muito bem em muitas situações do mundo real, como a famosa classificação de documentos e filtragem de spam. Eles requerem uma pequena quantidade de dados de treinamento para estimar os parâmetros necessários.
